```python
#12.

# Prompt the user to enter the starting and ending numbers for the prime number range
r1 = int(input('Enter the number from where you need the prime numbers: '))
r2 = int(input('Enter the number up to where you need the prime numbers: '))

# Iterate through the range from 'r1' to 'r2'
for num in range(r1, r2 + 1):
    # Prime numbers have exactly two factors: 1 and themselves
    # So, we need to check if a number has only two divisors.
    divisors = 0  # Initialize a divisor counter
    
    # Iterate through potential divisors from 1 to 'num'
    for divisor in range(1, num + 1):
        if num % divisor == 0:
            divisors += 1  # Increment the divisor counter
    
    # If a number has exactly two divisors, it's prime
    if divisors == 2:
        print(num)
```

```python
#14.

# Sample 2D list
matrix = [
    [12, 45, 23],
    [5, 67, 32],
    [56, 8, 91]
]

# Initialize a variable to store the maximum element, initially set to negative infinity
max_element = float("-inf")

# Nested for loops to iterate through the elements of the 2D list
for i in matrix:
    for j in i:
        # Check if the current element is greater than the current maximum
        if j > max_element:
            # Update the maximum element if the current element is greater
            max_element = j

# Print the maximum element found in the 2D list
print("The maximum element in the 2D list is:", max_element)
```

```python
#20.

# Function to calculate the GCD of two numbers
def find_gcd(a, b):
    # Find the minimum of the two numbers
    min_num = min(a, b)

    # Initialize a variable to store the GCD
    gcd = 1

    # Iterate from 1 to the minimum number
    for i in range(1, min_num + 1):
        if a % i == 0 and b % i == 0:
            # If 'i' is a common divisor, update the GCD
            gcd = i

    return gcd

# Input two numbers from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate the GCD using the find_gcd function
gcd = find_gcd(num1, num2)

# Print the GCD
print("The greatest common divisor (GCD) of", num1, "and", num2, "is", gcd)
```

```python
#31.

# Input the upper limit from the user
limit = int(input("Enter the upper limit for prime numbers: "))

# Create a list of boolean values to represent prime numbers (True initially)
is_prime = [True] * (limit + 1)

# Set the values for 0 and 1 to False as they are not prime
is_prime[0] = is_prime[1] = False

# Use the Sieve of Eratosthenes algorithm to mark non-prime numbers as False
for num in range(2, int(limit ** 0.5) + 1):
    if is_prime[num]:
        for multiple in range(num * num, limit + 1, num):
            is_prime[multiple] = False

# Generate a list of prime numbers using list comprehension
prime_numbers = [num for num in range(2, limit + 1) if is_prime[num]]

# Print the list of prime numbers
print("Prime Numbers up to", limit, ":", prime_numbers)
```

```python
#35.

from math import factorial

# Input the number of rows for Pascal's triangle from the user
num_rows = int(input("Enter the number of rows for Pascal's triangle: "))

# Generate Pascal's triangle using list comprehension
pascals_triangle = [[factorial(n) // (factorial(r) * factorial(n - r)) for r in range(n + 1)] for n in range(num_rows)]

# Print Pascal's triangle
for row in pascals_triangle:
    print(' '.join(map(str, row)))
```

```python
#37.

# Input a sentence from the user
sentence = input("Enter a sentence: ")

# Split the sentence into words
words = sentence.split()

# Initialize variables to keep track of the longest word and its length
longest_word = ""
max_length = 0

# Iterate through the words in the sentence
for word in words:
    # Remove punctuation from the word
    cleaned_word = ''.join(char for char in word if char.isalnum())

    # Check if the cleaned word is longer than the current longest word
    if len(cleaned_word) > max_length:
        longest_word = cleaned_word
        max_length = len(cleaned_word)

# Print the longest word
print("Longest Word in the Sentence:", longest_word)
```
